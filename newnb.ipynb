Determine the hour of the day with the highest number of trips.

```python
SELECT EXTRACT(HOUR FROM __time) AS hour_of_day,
       COUNT(*) AS num_trips
FROM yellowdata
GROUP BY EXTRACT(HOUR FROM __time)
ORDER BY num_trips DESC
LIMIT 1
```

```python
SELECT
    EXTRACT(YEAR FROM __time),
    EXTRACT(MONTH FROM __time),
    SUM(fare_amount + tip_amount)
FROM json
GROUP BY
    EXTRACT(YEAR FROM __time),
    EXTRACT(MONTH FROM __time)
ORDER BY
    EXTRACT(YEAR FROM __time),
    EXTRACT(MONTH FROM __time)
```

Rush by day of Month



```python
SELECT
    EXTRACT(DAY FROM __time) AS day_of_month,
    COUNT(*) AS trip_count
FROM ydata
GROUP BY EXTRACT(DAY FROM __time)
ORDER BY day_of_month
```
